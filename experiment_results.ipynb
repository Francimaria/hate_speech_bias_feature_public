```python
CLASSIFIERS = ["LR","DT","SVM","XGB","MLP","RF"]
FEATURES =  ['tf', 'tfidf',"glove", "fastText","bert",  'roberta']
METRICS = ['accuracy', 'auc', 'f1_score','fned', 'fped', 'subgroup', 'bpsn', 'bnsp']
```

# Average results

```python
def get_result_metric(features, metric, df):
    df_results = pd.DataFrame(columns=["model"])
    for clf in CLASSIFIERS:
        aux_c = df[df["model"].str.contains(clf)][["model",f"{metric}_mean", f"{metric}_std"]]
        
        aux_c[clf] =  aux_c[f"{metric}_mean"]
        aux_c = aux_c[["model", clf]]
        new_row =  {"model": [clf]}

        for f in features:
            new_row[f] = aux_c[aux_c["model"] == f"{clf}_{f}"][clf].values
        
        df_results = pd.concat([df_results, pd.DataFrame(new_row)], ignore_index=True)
    
    #transpose results
    df_results = df_results.T
    cols = df_results.iloc[0]
    df_results.columns = cols
    return df_results[1:]    
```

```python
dataset_name = "HE"
path = f"results/{dataset_name}/classifiers/metrics_average.csv"
df = pd.read_csv(path)

METRICS = ['auc', 'f1_score', 'fned', 'fped', 'subgroup']

for m in METRICS:
    print("----", m, "-----\n")
    results = get_result_metric(FEATURES, m, df)
    display(results)
```
