# Create k-folds





```python
def kfold(X, y, kf, path="dataset/", save=True):
    fold = 0
    X_train, X_test, X_val = {},{},{}
    y_train, y_test, y_val = {},{},{}

    for train, test in kf.split(X, y):
        
        X_train_, X_test[fold] = X.iloc[train], X.iloc[test]
        y_train_, y_test[fold] = y.iloc[train], y.iloc[test]

        # split train set .. 15% to validation set
        X_train[fold], X_val[fold], y_train[fold], y_val[fold] = train_test_split(X_train_, 
                                                                                  y_train_,
                                                                                    test_size=0.15, 
                                                                                    stratify=y_train_)
        fold += 1
    
    if save:
        if not check_exist(path):
            create_dir(path)

        save_pickle(f"{path}train.pickle", X_train)
        save_pickle(f"{path}train_labels.pickle", y_train)

        save_pickle(f"{path}val.pickle", X_val)
        save_pickle(f"{path}val_labels.pickle", y_val)

        save_pickle(f"{path}test.pickle", X_test)
        save_pickle(f"{path}test_labels.pickle", y_test)
```

```python
def pre_processing_k_fold(dataset="WH"):
    print(dataset)
    data = pd.read_csv(f"dataset/{dataset}/{dataset.lower()}_data.csv")
    # Preprocessing 
    data_prep = data
    print("len before: ",len(data_prep))
    data_prep["text"] = pre_processing(data["text"])

    #drop text with len bigger than one
    data_prep = data_prep[data_prep["text"].str.len()>1]
    print("len after: ",len(data_prep))

    #split k-fold
    kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
    path = f"dataset/{dataset}/preprocess/"
    kfold(data_prep["text"],data_prep["label"], kf, path=path, save=True)
```

```python
#WH dataset
pre_processing_k_fold(dataset="WH")

#DV dataset
pre_processing_k_fold(dataset="DV")
```
